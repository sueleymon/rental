## Check for null values

```python
df.isnull().sum()
```

```python
# drop all the empty rows that don't contain totalRent 
df.dropna(subset=['totalRent'],inplace=True)
```

```python
# Show the percentage of missing values
def missing_values(df,norows):   # input by the df and the number of rows that you want to show
    total = df.isnull().sum().sort_values(ascending=False)
    percent = ((df.isnull().sum().sort_values(ascending=False)/df.shape[0])*100).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return(missing_data.head(norows))
```

```python
# I wanted to remove any columns if the values are over 50%
missing_values(df, 8)
```

## Remove any outlier values

```python
# I want to delete the unrealistic values before I check for mixed data types
df.describe()
```

```python
# Floor should have up to 20 floors there is -1 because of cellar apartments
df = df[(df['floor'] >= -1) & (df['floor'] <= 20)]
```

```python
# Check if it worked
df['floor'].value_counts()
```

```python
df['serviceCharge'].describe()
```

```python
# Anything over 1000 is removed
df = df[(df['serviceCharge'] < 1000)]
```

```python
# Check if it worked
df['serviceCharge'].describe()
```

```python
df['livingSpace'].describe()
```

```python
# No living space is going to be counted that is under 10sqm and over 500sqm, that's the parameter I set
df = df[(df['livingSpace'] > 10) & (df['livingSpace'] < 500)]
```

```python
# Check if it worked
df['livingSpace'].describe()
```

```python
# This code sets the parameter of the rents and difference in rent with less outliers
df = df[(df['baseRent'] > 200) & (df['baseRent'] < 7000)]
df = df[(df['totalRent'] > 200) & (df['totalRent'] < 8000)]
df = df[(df['totalRent'] > df['baseRent'])]
df = df[(df['totalRent'] - df['baseRent']) < 500]
```

```python
# Check if it worked
df['baseRent'].describe().round(2)
```

```python
# Check if it worked
df['totalRent'].describe().round(2)
```
